# Session 10: Missing Values and Duplicates

## Purpose

This notebook checks the dataset for missing values and duplicate rows before preprocessing and model development.

```python
import pandas as pd
```

```python
df = pd.read_csv("../data/student-mat.csv", sep=";")

print("Dataset loaded successfully.")
print("Shape:", df.shape)

df.head()
```

## Check Missing Values Per Column

```python
print("Missing per column:")
print(df.isna().sum())
```

## Check Total Missing Values

```python
total_missing = df.isna().sum().sum()

print("Total missing values:", total_missing)
```

## Missing Value Summary Table

```python
missing_summary = pd.DataFrame({
    "Column": df.columns,
    "Missing_Count": df.isna().sum().values,
    "Missing_Percent": (df.isna().mean().values * 100).round(2)
})

missing_summary
```

## Prompt-Engineered Explanation Result

The dataset was checked for missing values and duplicate rows before preprocessing and model development.

The missing-value check showed that all columns have zero missing values.

The duplicate-row check also showed zero duplicate rows.

Because every column has zero missing values, imputation is not needed. Since the dataset is already complete, adding imputed values could introduce unnecessary artificial information.

Because the duplicate-row count is zero, no rows need to be removed.

## Recommendation

No cleaning action is required for missing values or duplicate rows at this stage. The dataset can move forward to preprocessing and feature engineering.
