```python
from pathlib import Path
import pandas as pd

PROJECT_ROOT = Path.cwd().parent

FULL_INFO_FILE = (
    PROJECT_ROOT
    / "data"
    / "processed"
    / "student_full_information.csv"
)

X_full = pd.read_csv(FULL_INFO_FILE)

print("X_full shape:", X_full.shape)
X_full.head()
```

```python
leak_cols = [
    c for c in X_full.columns
    if c in ("G1", "G2")
]

X_early = X_full.drop(columns=leak_cols)

print("Early-warning features:", X_early.shape[1])
```

```python
print("G1 present:", "G1" in X_early.columns)
print("G2 present:", "G2" in X_early.columns)
```

```python
print("Full-information shape:", X_full.shape)
print("Early-warning shape:", X_early.shape)
```

```python
assert "G1" not in X_early.columns
assert "G2" not in X_early.columns

print("Validation passed.")
```

```python
OUTPUT_FILE = (
    PROJECT_ROOT
    / "data"
    / "processed"
    / "student_early_warning.csv"
)

X_early.to_csv(
    OUTPUT_FILE,
    index=False
)

print("Saved:", OUTPUT_FILE)
```

# Session 21: Early-Warning Dataset Design

## Objective

Create an early-warning feature set by removing G1 and G2 from the full-information dataset.

## Results

- Full-information shape: (395, 41)
- Early-warning shape: (395, 39)

## Interpretation

The early-warning dataset excludes prior grades G1 and G2. These variables are highly predictive of final academic performance but may not be available early enough in the academic term for intervention purposes.

Removing G1 and G2 creates a more realistic early-warning scenario where predictions are made using demographic, family, behavioral, and school-related information available before major grades are recorded.
