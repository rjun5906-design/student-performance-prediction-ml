Project context:
We are predicting student performance using the UCI Student Performance Dataset. The final grade column is G3. We created a binary classification target called passed, where passed = 1 if G3 >= 10 and passed = 0 if G3 < 10.


# Target Definition Note

## Regression Target

The regression target is `G3`, which represents the student's final grade. This target will be used when the model predicts a numeric final grade.

## Classification Target

The classification target is `passed`. This target converts the final grade into a binary outcome.

The classification rule is:

passed = 1 if G3 >= 10

passed = 0 if G3 < 10

This means that students with final grades of 10 or higher are labeled as passing, while students with final grades below 10 are labeled as not passing.

## Chosen Threshold

The chosen threshold is `G3 >= 10` for passing. Students with `G3 < 10` are considered at risk or not passing.

## Limitation

One limitation of this threshold is that it creates a hard cutoff. For example, a student with `G3 = 9` is labeled as not passing, while a student with `G3 = 10` is labeled as passing. However, these two students may have very similar academic performance.

## Final Decision

For the first version of the project, we will keep the `G3 >= 10` threshold because it is simple, clear, and easy to explain. However, we will document its limitation and may compare it with other thresholds, such as `G3 < 12`, in a later version of the project.
