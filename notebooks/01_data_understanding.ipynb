```python
import pandas as pd

# Load the UCI Student Performance mathematics dataset
# The CSV file uses semicolon ; as the separator

df = pd.read_csv("student-mat.csv", sep=";")

print("Dataset shape:", df.shape)

df.head()
```

# CSV Loading Explanation
Summary

The CSV file appears to have loaded correctly because the dataset shape is (395, 33), indicating that the file contains 395 student records and 33 variables. The output from df.head() also displays the first five rows of the dataset, confirming that the data was successfully loaded into a pandas DataFrame.

Interpretation

Two useful checks for confirming that the CSV loaded correctly are df.shape and df.head(). The first check verifies the number of rows and columns in the dataset. The second check displays the first few rows so that the column names and values can be visually inspected. The separator is important because the UCI student-mat.csv file uses semicolons rather than commas. If the wrong separator is used, pandas may load the entire row into a single column instead of splitting the dataset into 33 columns.
