```python
df = pd.read_csv("../data/student-mat.csv", sep=";")
```

```python
import os
import matplotlib.pyplot as plt
import seaborn as sns

os.makedirs("figures", exist_ok=True)

mean_grades = (
    df.groupby("studytime")["G3"]
      .mean()
      .round(2)
)

print("Mean final grade by studytime:")
print(mean_grades)

plt.figure(figsize=(7,5))

sns.boxplot(
    x="studytime",
    y="G3",
    data=df
)

plt.title("Final Grade by Study Time")
plt.xlabel("Study Time Group")
plt.ylabel("Final Grade (G3)")

plt.savefig(
    "figures/studytime_vs_grade.png",
    dpi=300,
    bbox_inches="tight"
)

plt.show()
```

## Session 14 Findings

Students with higher study-time levels generally tend to earn higher final grades.

However, there is overlap between groups, suggesting that study time alone does not completely explain student performance.

Study time may be useful as a predictor, but it should be combined with other variables such as absences and past failures.

# Session 15: Attendance, Failures, and Performance Analysis

```python
import os
import matplotlib.pyplot as plt
import seaborn as sns

os.makedirs("figures", exist_ok=True)

print("Mean final grade by failures:")

failure_means = (
    df.groupby("failures")["G3"]
      .mean()
      .round(2)
)

print(failure_means)

plt.figure(figsize=(8,5))

sns.scatterplot(
    x="absences",
    y="G3",
    data=df,
    alpha=0.5
)

plt.title("Final Grade vs Absences")
plt.xlabel("Absences")
plt.ylabel("Final Grade (G3)")

plt.savefig(
    "figures/absences_vs_grade.png",
    dpi=300,
    bbox_inches="tight"
)

plt.show()
```
