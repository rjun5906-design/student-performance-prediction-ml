```python
import pandas as pd

df = pd.read_csv("data/student-mat.csv", sep=";")

num_cols = df.select_dtypes(include="number").columns.tolist()

numeric_summary = df[num_cols].agg(["min", "max", "mean"]).T
numeric_summary = numeric_summary.round(2)

display(numeric_summary)
```

```python
import pandas as pd

df = pd.read_csv("../data/student-mat.csv", sep=";")

print(df.shape)
df.head()
```

```python
num_cols = df.select_dtypes(include="number").columns.tolist()

numeric_summary = (
    df[num_cols]
    .agg(["min", "max", "mean"])
    .T
    .round(2)
)

display(numeric_summary)
```

```python
import os

os.makedirs("../reports", exist_ok=True)

numeric_summary.to_csv(
    "../reports/session12_numeric_summary.csv"
)

print("Saved Session 12 numeric summary.")
```
