```python
from pathlib import Path
import pandas as pd

PROJECT_ROOT = Path.cwd().parent

INPUT_FILE = PROJECT_ROOT / "data" / "student-mat.csv"

df = pd.read_csv(INPUT_FILE, sep=";")

print("Dataset shape:", df.shape)
df.head()
```

```python
cat_cols = df.select_dtypes(include="object").columns

print("Categorical columns:")
print(list(cat_cols))
print("Number of categorical columns:", len(cat_cols))
```

```python
df_encoded = pd.get_dummies(
    df,
    columns=list(cat_cols),
    drop_first=True
)

print("Original shape:", df.shape)
print("Encoded shape:", df_encoded.shape)
```

# Session 19: Feature Engineering I

## One-Hot Encoding Results

### Original Shape
(395, 33)

### Encoded Shape
(395, 42)

### Summary

The dataset originally contained 395 student records and 33 variables. Seventeen categorical variables were identified and transformed into dummy variables using one-hot encoding. After encoding, the dataset contained 42 columns while preserving all 395 student records.

### Interpretation

The increase from 33 to 42 columns occurred because categorical variables were expanded into binary indicator variables. Variables with multiple categories generated multiple dummy columns. The parameter `drop_first=True` was used to remove one category from each categorical variable, making that category the reference category and reducing redundancy.

The row count remained unchanged because one-hot encoding transforms existing variables rather than adding or removing observations. Each row still represents the same student.

### Recommendation

Verify that:
- The row count remains 395 before and after encoding.
- No object-type columns remain.
- The target variable G3 is unchanged.
- Dummy variables contain only 0 and 1 values.
- No duplicate columns were introduced.

The encoded dataset is now ready for feature scaling and machine-learning model development.

```python
from pathlib import Path

PROJECT_ROOT = Path.cwd().parent

OUTPUT_FILE = (
    PROJECT_ROOT
    / "data"
    / "processed"
    / "student_data_encoded.csv"
)

df_encoded.to_csv(
    OUTPUT_FILE,
    index=False
)

print("Saved:", OUTPUT_FILE)
```
